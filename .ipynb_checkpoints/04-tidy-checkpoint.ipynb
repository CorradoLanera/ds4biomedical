```python
import pandas as pd
import numpy as np
```

```python
tumor = pd.read_csv("data/tumorgrowth_long.csv")
tumor
```

```python
tumor_tidy = tumor.melt(
    id_vars = ["Group", "Grp", "ID"],
    var_name = "day",
    value_name = "size"
)
```

```python
avg_mean = (tumor_tidy
    .groupby(["Group", "day"])
    .agg(avg = ("size", "mean"))
    .reset_index()
    .assign(day = pd.to_numeric(tumor_tidy["day"]))
    .sort_values(["Group", "day"])
)
avg_mean
```
