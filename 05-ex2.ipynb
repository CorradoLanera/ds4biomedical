```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
blood = pd.read_csv("data/medicaldata_blood_storage.csv")
blood
```

```python
blood.columns
```

```python
blood.info()
```

```python
blood_recurrence_count = (
    blood
    .groupby("Recurrence")
    .agg(count_recurrence = ("Recurrence", "count"))
    .reset_index()
)
```

```python
blood["Recurrence"].value_counts()  # class imbalance!
```

```python
blood_recurrence_count
```

```python
sns.countplot(x = "Recurrence", data = blood)
```

```python
sns.barplot(x = "Recurrence", y = "count_recurrence", data = blood_recurrence_count)
```

```python
sns.histogram(data = blood, x = "Age")
```

```python
sns.boxplot(x = "TVol", y = "Age", data = blood)
```

```python
sns.violinplot(x = "TVol", y = "Age", data = blood)
```

```python
sns.lmplot(x = "PVol", y = "PreopPSA", data = blood,
           hue = "sGS",
           row = "RBC.Age.Group",
           col = "Recurrence"
          )
```

```python
fi, ax = plt.subplots(1, 1)
```

```python
sns.regplot(x = "PVol", y = "PreopPSA", data = blood, ax = ax)
```
